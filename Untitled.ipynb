```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sstats
import numpy as np 
```

```python
def parse_money(x):
    damage = None
    try:
        damage = int(x)
        if damage == 0:
            return damage
    except:
        try:
            scale = x[-1]
            if len(x) == 1 and x == 'K':
                return 1000
            elif scale == 'K':
                damage = float(x[:-1])
                return damage * 1000
            elif scale == 'M':
                damage = float(x[:-1])
                return damage * 1000000
        except:
            print x
            raise
```

```python
df = pd.read_csv('data_tables/storm_details.csv', 
                 index_col=0,
                 header = 0,
                 usecols=['EVENT_ID', 'STATE','YEAR','MONTH_NAME',
                          'EVENT_TYPE','BEGIN_DATE_TIME',
                          'END_DATE_TIME','INJURIES_DIRECT','INJURIES_INDIRECT','DEATHS_DIRECT','DEATHS_INDIRECT',
                          'DAMAGE_PROPERTY',
                          'TOR_F_SCALE','TOR_LENGTH','TOR_WIDTH','BEGIN_LAT','BEGIN_LON','END_LAT','END_LON'])
                #converters= {#'DAMAGE_PROPERTY': parse_money,
                             #'DAMAGE_CROPS': parse_money,
                             #'BEGIN_DATE_TIME': pd.to_datetime,
                             #'#END_DATE_TIME': pd.to_datetime})
df.columns
```

```python
t_df = df[(df['EVENT_TYPE']=='Tornado') & (df['YEAR']>2007) & (df['TOR_WIDTH']>0) &(df['TOR_LENGTH']>0)].dropna()
t_df.BEGIN_DATE_TIME = pd.to_datetime(t_df.BEGIN_DATE_TIME)
t_df.END_DATE_TIME = pd.to_datetime(t_df.END_DATE_TIME)
t_df['DURATION_M'] = (t_df.END_DATE_TIME-pd.to_datetime(t_df.BEGIN_DATE_TIME))/np.timedelta64(1,'m')
t_df.DAMAGE_PROPERTY = t_df.DAMAGE_PROPERTY.apply(parse_money)
t_df.dropna(inplace=True)
t_df = t_df[(t_df['DURATION_M']>0) & (t_df.DURATION_M<1000)]
t_df["START_TIME"]
print t_df.shape
t_df.dtypes
```
